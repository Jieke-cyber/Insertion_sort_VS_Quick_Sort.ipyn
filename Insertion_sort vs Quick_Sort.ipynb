# Insertion-sort VS Quick-sort

## introduzione

*Insertion-sort* è un **algoritmo di ordinameto**, molto efficiente per dati di dimensione piccole o parzialmente oridinati, *Quick-sort*, invece, è un algoritmo più soffistica ed efficiente anche per dati di grande dimensoine.

## Insertion-sort

### Descrizione

*Insertion-sort* è un algoritmo di ordinamento semplice e intuitivo che costruisce gradualmente una lista ordinata. Ogni volta che si inserisce un nuovo elemento, **si individua la posizione corretta a partire dal primo elemento della lista** per collocare tale elemento all'interno della lista già **ordinata**.

## Codice

```python
def insertion_sort(arr):
    # Itera dall'indice 1 fino alla fine dell'array
    for i in range(1, len(arr)):
        key = arr[i]  # Elemento da inserire
        j = i - 1
        
        # Sposta gli elementi di arr[0..i-1] che sono maggiori di key
        # una posizione avanti rispetto alla loro posizione attuale
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        
        # Inserisce l'elemento in posizione corretta
        arr[j + 1] = key
```

## Esempio di utilizzo

```python
data = [12, 11, 13, 5, 6]
insertion_sort(data)
print("Array ordinato:", data)
```

## Tempo di esecuzione
**Caso migliore**: la lista è già ordinata, ha complessità temporale __O(n)__

**Caso peggiore**: la lista è ordinata in modo decrescente, ha complessità temporale __O($n^2$)__

# Quick-sort

## Descrizione

Quick-sort ordina la lista **suddividendola**, in corrispondenza di un indice chiamato ***pivot***, in due **sottoliste**: una contenente gli elementi più piccoli e l'altra gli elementi più grandi del pivot. Questo processo viene ripetuto in modo **ricorsivo** fino a ottenere l'intera lista ordinata.

## Codice

```python
def partition(arr, low, high):
    # Usiamo l'ultimo elemento come pivot
    pivot = arr[high]
    i = low - 1  # Indice per tenere traccia della "frontiera" degli elementi minori del pivot

    # Spostiamo gli elementi in base al pivot
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1  # Spostiamo la frontiera
            arr[i], arr[j] = arr[j], arr[i]  # Scambiamo gli elementi

    # Posizioniamo il pivot nella posizione corretta
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1  # Restituiamo l'indice del pivot

def quick_sort(arr, low, high):
    if low < high:
        # Otteniamo l'indice del pivot dopo la partizione
        pivot_index = partition(arr, low, high)

        # Applichiamo quick sort a ciascuna metà della lista
        quick_sort(arr, low, pivot_index - 1)
        quick_sort(arr, pivot_index + 1, high)
```

La funzione ***partition*** è finalizzata a dividere la lista in due sottoliste che contiene rispettivamente gli elementi **minori** e gli elementi **maggiori** al pivot, e infine **ritorna** la posizione del pivot nella lista

## Esempio di utilizzo

```python
# Esempio di utilizzo
data = [12, 7, 14, 9, 10, 11]
quick_sort(data, 0, len(data) - 1)
print("Array ordinato:", data)
```

## Colcusione

**Test con lista casuale**: *Quick-sort* mantiene un tempo di esecuzione molto basso e quasi costante, mostrando la sua efficienza nel gestire liste non ordinate grazie alla sua complessità *O(nlogn)*. *Insertion-sort*, invece, presenta un incremento marcato nei tempi di esecuzione, dovuto alla sua complessità *O($n^2$)*, che diventa sempre più evidente all’aumentare del numero di elementi.

**Test con lista crescente**: *Quick-sort* mostra un forte aumento nei tempi di esecuzione. Questo è dovuto al fatto che una lista ordinata rappresenta uno scenario sfavorevole per *Quick-sort*, il quale può incorrere nel caso peggiore *O($n^2$)* se non è ottimizzato per questo tipo di input. *Insertion-sort*, d’altra parte, è molto efficiente con liste già ordinate, mantenendo un tempo di esecuzione quasi costante, poiché richiede poche operazioni di scambio e confronto in questo scenario.

**Test con lista decrescente**: Su una lista ordinata in modo decrescente, *Quick-sort* risulta più efficiente rispetto alla lista ordinata crescente, mostrando un aumento dei tempi di esecuzione più contenuto. *Insertion-sort*, invece, continua a incrementare rapidamente con la dimensione della lista, mostrando i limiti della sua complessità *O($n^2$)* quando l'input è inversamente ordinato rispetto all’ordine finale desiderato.

*Quick-sort* è particolarmente **efficiente** per liste casuali e mantiene una buona performance su liste decrescenti, dimostrando la sua adattabilità nella maggior parte dei casi reali. *Insertion-sort* è **vantaggioso** solo su liste già ordinate, ma diventa rapidamente **inefficiente** con liste di grandi dimensioni o non ordinate.

## Tempo di esecuzione

**Caso medio**: Nella maggior parte dei casi, Quick Sort ha una complessità temporale di *O(nlogn)*, che si ottiene quando il pivot divide in modo equilibrato la lista in due sottoliste quasi uguali a ogni livello di ricorsione. Questo è il caso medio, che si verifica per una lista con elementi distribuiti casualmente.

**Caso migliore**: Il caso migliore si verifica se il pivot selezionato è sempre l'elemento medio (teorico) della lista, che produce partizioni perfettamente bilanciate. In pratica, questo caso è simile al caso medio ed ha complessità *O(nlogn)*.

**Caso peggiore**: Il caso peggiore con complessità *O($n^2$)* si verifica quando il pivot è sempre l'elemento più piccolo o più grande della lista. Questo accade, ad esempio, con un array già ordinato (o ordinato in ordine inverso) quando si sceglie sempre il primo o l'ultimo elemento come pivot.

## Testing

```python
import random
from timeit import default_timer as timer
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.setrecursionlimit(10 ** 9)

# Funzioni per generare liste
def listaCasuale(num_elementi):
    return [random.random() for _ in range(num_elementi)]

def listaOrdinata(num_elementi):
    return sorted([random.random() for _ in range(num_elementi)])

def listaOrdinataDecrescente(num_elementi):
    return sorted([random.random() for _ in range(num_elementi)], reverse=True)

# Funzione di test
def test(alg_lista, num_elementi, iterazioni):
    dimensioni_lista = []
    insertion_sort_t = []
    quicksort_t = []

    for i in range(iterazioni):
        n = (num_elementi) * (i + 1)
        dimensioni_lista.append(n)
        print(f"Numero di elementi: {n}")
        
        # Genera una nuova lista per ogni iterazione
        lista = alg_lista(n)
        copia_lista = lista.copy()

        # Testa Insertion Sort
        start = timer()
        insertion_sort(lista)
        stop = timer()
        insertion_sort_t.append(stop - start)
        
        # Testa Quicksort
        start = timer()
        quick_sort(copia_lista, 0, n-1)
        stop = timer()
        quicksort_t.append(stop - start)

    # Crea il grafico che mostra il confronto temporale
    plt.plot(dimensioni_lista, insertion_sort_t, label='Insertion Sort')
    plt.plot(dimensioni_lista, quicksort_t, label='Quicksort')
    plt.xlabel('Numero di elementi')
    plt.ylabel('Tempo di esecuzione (secondi)')
    plt.title('Confronto tra Insertion Sort e Quicksort')
    plt.legend()
    plt.show()

# Funzione principale
def main():
    num_elementi = 1700
    iterazioni = 15
    print('Test con lista casuale...')
    test(listaCasuale, num_elementi, iterazioni)
    print('Test con lista ordinata...')
    test(listaOrdinata, num_elementi, iterazioni)
    print('Test con lista ordinata in ordine decrescente...')
    test(listaOrdinataDecrescente, num_elementi, iterazioni)

main()
```
